```python
def combine_datasets(*datasets):
    combined = []
    for dataset in datasets:
        combined.extend(dataset)
    return combined

ds1 = [{"id": 1, "value": 10}, {"id": 2, "value": 20}]
ds2 = [{"id": 3, "value": 30}, {"id": 4, "value": 40}]
ds3 = [{"id": 5, "value": 50}]

print(combine_datasets(ds1, ds2, ds3))
```
